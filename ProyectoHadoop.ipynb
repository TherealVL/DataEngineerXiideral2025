# Proyecto Hadoop
Este proyecto utiliza los servicios de AWS con Hadoop para procesar grandes archivos de gran volumen.

*OBJETIVOS*


1.   Obtener el total de GET /image/
2.   Obtener el total de POST
3.   Obtener el total de error





Copiar archivo access.log.bz2 de S3

```python
aws s3 cp s3://proyecto-final-spark-hadoop-data-engineer/access.log.bz2 /root/input/access.log.bz2
```

s3.png

Listamos para comprobar la creación del archivo

```python
ls
```

21.png

Copiar el archivo access.log.bz2 a Hadoop

```python
hdfs dfs -cp input/access.log.bz2 input/access.log.bz2
hdfs dfs -ls input
```

22.png



Ejecutar un trabajo en HADOOP para "GET /image/".

```python
hadoop jar /usr/lib/hadoop-mapreduce/hadoop-mapreduce-examples.jar grep input/access.log.bz2 salida_get_logs_luis "GET /image/"
hdfs dfs -ls salida_get_logs_luis
```

24.png

Resultado de nuestra consulta "GET"

```python
hdfs dfs -cat salida_get_logs_luis/part-r-00000
```
